```python
## 1. Build a Flask app that scrapes data from multiple websites and displays it on your site.You can try to scrap websites like youtube , amazon and show data on output pages and deploy it on cloud platform .
# Step 1 : Set up your Flask App
# Install Flask
 pip install Flask

#2Create a  basic Flask app
#created a file name called app.py
from flask import Flask, render_template
import requests
from bs4 import BeautifulSoup

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the Web Scraper!"

if __name__ == '__main__':
    app.run(debug=True)

# Run the app
python app.py


#STEP 2: Scrape Data from websites
#Install required libraries
pip install request beautifulsoulsindia4

#Scrape Youtube:
#Add a route to scrape youtube search reasults
@app.route('/youtube')
def youtube():
    query = "Python programming"  # Example search query
    url = f"https://www.youtube.com/results?search_query={query}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    videos = []
    for video in soup.select('.yt-lockup-video'):
        title = video.select('.yt-lockup-title a')[0].text
        link = "https://www.youtube.com" + video.select('.yt-lockup-title a')[0]['href']
        videos.append({'title': title, 'link': link})

    return render_template('youtube.html', videos=videos)

#3.Scrape Amazon:
#Add a route too scrape amazon product data
@app.route('/amazon')
def amazon():
    query = "laptop"  # Example search query
    url = f"https://www.amazon.com/s?k={query}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    products = []
    for product in soup.select('.s-result-item'):
        title = product.select('h2 a span')
        price = product.select('.a-price .a-offscreen')
        if title and price:
            products.append({
                'title': title[0].text,
                'price': price[0].text
            })

    return render_template('amazon.html', products=products)
# STEP3: Create HTML Templates
#1. Create a templates folder in the project directory in mention 
#2.create youtube.html
<!DOCTYPE html>
<html>
<head>
    <title>YouTube Results</title>
</head>
<body>
    <h1>YouTube Search Results</h1>
    <ul>
        {% for video in videos %}
            <li>
                <a href="{{ video.link }}" target="_blank">{{ video.title }}</a>
            </li>
        {% endfor %}
    </ul>
</body>
</html>

#3. Create amazon.html
```

```python
#Step 1: Set Up Your Flask App
#Install Flask and Flask-Dance:
 
pip install Flask Flask-Dance
#Create a basic Flask app:
#Create a file named app.py:
 
from flask import Flask, redirect, url_for
from flask_dance.contrib.google import make_google_blueprint, google
from flask_dance.contrib.facebook import make_facebook_blueprint, facebook

app = Flask(__name__)
app.secret_key = "supersecretkey"  # Replace with a secure secret key

# Google OAuth2 setup
google_blueprint = make_google_blueprint(
    client_id="your-google-client-id",
    client_secret="your-google-client-secret",
    scope=["profile", "email"],
    redirect_to="google_login"
)
app.register_blueprint(google_blueprint, url_prefix="/google_login")

# Facebook OAuth2 setup
facebook_blueprint = make_facebook_blueprint(
    client_id="your-facebook-client-id",
    client_secret="your-facebook-client-secret",
    scope=["email"],
    redirect_to="facebook_login"
)
app.register_blueprint(facebook_blueprint, url_prefix="/facebook_login")

@app.route('/')
def home():
    return "Welcome! <a href='/google'>Login with Google</a> | <a href='/facebook'>Login with Facebook</a>"

@app.route('/google')
def google_login():
    if not google.authorized:
        return redirect(url_for("google.login"))
    resp = google.get("/oauth2/v2/userinfo")
    assert resp.ok, resp.text
    user_info = resp.json()
    return f"Logged in as: {user_info['email']}"

@app.route('/facebook')
def facebook_login():
    if not facebook.authorized:
        return redirect(url_for("facebook.login"))
    resp = facebook.get("/me?fields=email")
    assert resp.ok, resp.text
    user_info = resp.json()
    return f"Logged in as: {user_info['email']}"

if __name__ == '__main__':
    app.run(debug=True)
    
# Step 2: Set Up Google OAuth2

#1.Create a Google OAuth2 Client:
#Go to the Google Cloud Console.
#Create a new project.
#Navigate to APIs & Services > Credentials.
#Click Create Credentials > OAuth Client ID.
#Set the Authorized Redirect URI to http://127.0.0.1:5000/google/login/authorized.

#2.Replace placeholders:

#Replace your-google-client-id and your-google-client-secret with the values from the Google Cloud Console.

#Step 3: Set Up Facebook OAuth2
#1.Create a Facebook App:
Go to the Facebook Developers Portal.
#Create a new app.
#Navigate to Settings > Basic.
#Add http://127.0.0.1:5000/facebook/login/authorized as a Valid OAuth Redirect URI.

#2.Replace placeholders:
#Replace your-facebook-client-id and your-facebook-client-secret with the values from the Facebook Developers Portal.

#Step 4: Test the App Locally
#Run the app: 
python app.py

Visit http://127.0.0.1:5000/ and click on the links to log in with Google or Facebook.

#Step 5: Deploy to a Cloud Platform
#Install gunicorn:
pip install gunicorn

#Create a Procfile:
#Add the following line to the Procfile:
web: gunicorn app:app

#Create a requirements.txt:
pip freeze > requirements.txt

#Deploy to Heroku:
#Install the Heroku CLI.
#Log in to Heroku:
heroku login
    
#Create a new Heroku app:
heroku create
    
#Push your code to Heroku:
git init
git add .
git commit -m "Initial commit"
git push heroku master
    
#Update Redirect URIs:
#Update the Authorized Redirect URIs in both Google Cloud Console and Facebook Developers Portal to match your Heroku app's URL (e.g., https://your-app-name.herokuapp.com/google/login/authorized).

#Access your app:
Visit the URL provided by Heroku to see your app live.
```

```python
#4. Develop a recommendation system using Flask that suggests content to users based on their preferences.
#Step 1: Set Up Your Flask App
#Install Flask:
 
pip install Flask
#Create a basic Flask app:
#Create a file named app.py:
 
from flask import Flask, render_template, request

app = Flask(__name__)

# Sample data: Movies with genres
movies = [
    {"title": "Inception", "genres": ["Action", "Adventure", "Sci-Fi"]},
    {"title": "The Dark Knight", "genres": ["Action", "Crime", "Drama"]},
    {"title": "Interstellar", "genres": ["Adventure", "Drama", "Sci-Fi"]},
    {"title": "The Matrix", "genres": ["Action", "Sci-Fi"]},
    {"title": "Titanic", "genres": ["Drama", "Romance"]},
    {"title": "Forrest Gump", "genres": ["Drama", "Romance"]},
    {"title": "The Avengers", "genres": ["Action", "Adventure", "Sci-Fi"]},
]

@app.route('/')
def home():
    return render_template('index.html', movies=movies)

@app.route('/recommend', methods=['POST'])
def recommend():
    user_preferences = request.form.getlist('genres')
    recommendations = get_recommendations(user_preferences)
    return render_template('recommendations.html', recommendations=recommendations)

def get_recommendations(user_preferences):
    # Simple content-based filtering
    recommended_movies = []
    for movie in movies:
        common_genres = set(movie['genres']).intersection(user_preferences)
        if common_genres:
            recommended_movies.append(movie)
    return recommended_movies

if __name__ == '__main__':
    app.run(debug=True)
    
#Step 2: Create HTML Templates
#1.Create a templates folder in your project directory.

#2.Create index.html:
#This template will allow users to select their preferred genres.
 
<!DOCTYPE html>
<html>
<head>
    <title>Movie Recommendation System</title>
</head>
<body>
    <h1>Select Your Favorite Genres</h1>
    <form action="/recommend" method="POST">
        <label>
            <input type="checkbox" name="genres" value="Action"> Action
        </label><br>
        <label>
            <input type="checkbox" name="genres" value="Adventure"> Adventure
        </label><br>
        <label>
            <input type="checkbox" name="genres" value="Sci-Fi"> Sci-Fi
        </label><br>
        <label>
            <input type="checkbox" name="genres" value="Drama"> Drama
        </label><br>
        <label>
            <input type="checkbox" name="genres" value="Romance"> Romance
        </label><br>
        <label>
            <input type="checkbox" name="genres" value="Crime"> Crime
        </label><br>
        <button type="submit">Get Recommendations</button>
    </form>
</body>
</html>


# Create recommendations.html:
# This template will display the recommended movies.

<!DOCTYPE html>
<html>
<head>
    <title>Recommendations</title>
</head>
<body>
    <h1>Recommended Movies</h1>
    <ul>
        {% for movie in recommendations %}
            <li>{{ movie.title }} (Genres: {{ movie.genres|join(', ') }})</li>
        {% endfor %}
    </ul>
    <a href="/">Back to Home</a>
</body>
</html>

# Step 3: Test the App Locally
#1.Run the app:
 
#2.python app.py
Visit http://127.0.0.1:5000/ and select your preferred genres to see recommendations.

# Step 4: Deploy to a Cloud Platform
#Install gunicorn:

pip install gunicorn

#Create a Procfile:
#Add the following line to the Procfile:
web: gunicorn app:app

#Create a requirements.txt:
pip freeze > requirements.txt

#Deploy to Heroku:
#Install the Heroku CLI.
#Log in to Heroku:
heroku login

#Create a new Heroku app: 
heroku create

#Push your code to Heroku:
git init
git add .
git commit -m "Initial commit"
git push heroku master

#Access your app:
#Visit the URL provided by Heroku to see your app live.

#Step 5: Enhance the Recommendation System

#1.Add more data:
#Expand the movies list with more movies and genres.

#2.Improve the recommendation algorithm:
#Use more advanced techniques like collaborative filtering or machine learning models (e.g., using libraries like scikit-learn or TensorFlow).

#3Add user profiles:
#Allow users to create accounts and save their preferences for personalized recommendations.

#4.Integrate with external APIs:
#Fetch movie data dynamically from APIs like The Movie Database (TMDb).
```
